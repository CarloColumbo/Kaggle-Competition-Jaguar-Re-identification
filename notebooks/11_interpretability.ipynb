## Interpretability

```python
class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer
        self.gradients = None
        self.activations = None
        self.hook_handles = []

        self._register_hooks()

    def _register_hooks(self):
        def forward_hook(module, input, output):
            self.activations = output.detach()
        def backward_hook(module, grad_in, grad_out):
            self.gradients = grad_out[0].detach()
        self.hook_handles.append(self.target_layer.register_forward_hook(forward_hook))
        self.hook_handles.append(self.target_layer.register_backward_hook(backward_hook))

    def __call__(self, x, class_idx=None):
        self.model.zero_grad()
        embedding = self.model(x)
        # If class_idx is None, take sum of embeddings for attribution
        if class_idx is None:
            score = embedding.sum()
        else:
            score = embedding[:, class_idx].sum()
        score.backward(retain_graph=True)
        weights = self.gradients.mean(dim=(2,3), keepdim=True)  # GAP over H,W
        cam = F.relu((weights * self.activations).sum(1, keepdim=True))
        cam = F.interpolate(cam, size=x.shape[2:], mode='bilinear', align_corners=False)
        cam = cam.squeeze().cpu().numpy()
        cam = (cam - cam.min()) / (cam.max() - cam.min() + 1e-6)  # normalize 0-1
        return cam

    def remove_hooks(self):
        for h in self.hook_handles:
            h.remove()
```

```python
def compute_embeddings(dataloader, model):
    model.eval()
    all_embeddings = []
    all_labels = []
    with torch.no_grad():
        for imgs, labels in tqdm(dataloader):
            imgs = imgs.to(device)
            emb = model(imgs)
            all_embeddings.append(emb.cpu())
            all_labels.append(labels)
    return torch.cat(all_embeddings), torch.cat(all_labels)
```

```python
def compute_map(embeddings, labels):
    sim_matrix = cosine_similarity(embeddings)
    labels = labels.cpu().numpy()
    np.fill_diagonal(sim_matrix, -1)

    query_aps = {}

    for query_idx in range(len(labels)):

        query_label = labels[query_idx]

        similarities = sim_matrix[query_idx]

        gallery_labels = labels.copy()

        is_match = (gallery_labels == query_label).astype(int)
        is_match[query_idx] = 0

        sorted_indices = np.argsort(-similarities)
        sorted_matches = is_match[sorted_indices]

        n_positives = sorted_matches.sum()

        if n_positives == 0:
            continue

        cumsum = np.cumsum(sorted_matches)

        precision_at_k = cumsum / np.arange(1, len(sorted_matches) + 1)

        ap = np.sum(precision_at_k * sorted_matches) / n_positives

        query_aps[query_idx] = (query_label, ap)

    identity_aps = {}

    for _, (label, ap) in query_aps.items():

        identity_aps.setdefault(label, []).append(ap)

    identity_mean_aps = [np.mean(v) for v in identity_aps.values()]

    balanced_map = np.mean(identity_mean_aps)

    return balanced_map
```

```python
import random

def mask_topk(img, cam, k=0.1, mask_value=0.5):
    _, _, H, W = img.shape
    cam_tensor = torch.tensor(cam).unsqueeze(0).unsqueeze(0)
    print(cam_tensor.shape)
    cam_upsampled = F.interpolate(cam_tensor, size=(H, W), mode='bilinear', align_corners=False)
    cam_upsampled = cam_upsampled.squeeze().numpy()  # now [H,W]
    
    # cam: HxW normalized 0-1
    threshold = np.percentile(cam_upsampled, 100*(1-k))
    mask = cam_upsampled >= threshold
    masked_img = img.clone()
    for c in range(3):
        masked_img[0,c][mask] = mask_value
    return masked_img

def mask_random(img, k=0.1, mask_value=0.5):
    _, H, W = img.shape
    num_pixels = int(H*W*k)
    masked_img = img.clone()
    for c in range(3):
        indices = random.sample(range(H*W), num_pixels)
        masked_img[0,c].view(-1)[indices] = mask_value
    return masked_img
```

```python
def interpretability_experiment(dataloader, model, gradcam, k=0.1):
    model.eval()
    all_embeddings_masked = []
    all_embeddings_random = []
    all_labels = []

    for imgs, labels in tqdm(dataloader):
        imgs = imgs.to(device)
        all_labels.append(labels)
        # Compute CAM
        cam = gradcam(imgs)
        # Mask top-k
        masked = torch.stack([mask_topk(imgs[i:i+1].cpu(), cam, k=k) for i in range(imgs.size(0))]).to(device)
        random_masked = torch.stack([mask_random(imgs[i:i+1].cpu(), k=k) for i in range(imgs.size(0))]).to(device)
        # Get embeddings
        emb_masked = model(masked)
        emb_random = model(random_masked)
        all_embeddings_masked.append(emb_masked.cpu())
        all_embeddings_random.append(emb_random.cpu())

    all_labels = torch.cat(all_labels)
    emb_masked = torch.cat(all_embeddings_masked)
    emb_random = torch.cat(all_embeddings_random)

    mAP_masked = compute_map(emb_masked, all_labels)
    mAP_random = compute_map(emb_random, all_labels)
    print(f"mAP drop top-{int(k*100)}% mask: {mAP_masked:.4f}")
    print(f"mAP drop random-{int(k*100)}% mask: {mAP_random:.4f}")
```

```python
def sanity_check(dataloader, projection, gradcam):
    # Randomize projection
    for layer in projection:
        if hasattr(layer, 'reset_parameters'):
            layer.reset_parameters()
    # Compare CAM visually
    imgs, _ = next(iter(dataloader))
    cam_rand = gradcam(imgs.to(device))
    cam_orig = gradcam(imgs.to(device))  # you may save original CAM first
    # Plot example
    plt.figure(figsize=(8,4))
    plt.subplot(1,2,1)
    plt.imshow(cam_orig)
    plt.title("Original CAM")
    plt.subplot(1,2,2)
    plt.imshow(cam_rand)
    plt.title("Randomized CAM")
    plt.show()
```

```python
class ImageDataset(Dataset):
    """
    PyTorch Dataset for images.
    """

    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform
    
    def __len__(self):
        return len(self.labels)
    
    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        img = Image.open(image_path).convert("RGB")
        label = self.labels[idx]
        if self.transform:
            img = self.transform(img)
        return img, label
```

```python
# Compute similarity for each pair
print("Computing pairwise similarities...")
similarities = []

if rerank:
    sim_matrix = -reranked_test_dist
else:
    # Compute cosine similarity matrix
    sim_matrix = cosine_similarity(test_finetuned_embeddings)
np.fill_diagonal(sim_matrix, 1) # self-similarity

print(f"Similarity matrix shape: {sim_matrix.shape}", f"Sim range: [{sim_matrix.min():.4f}, {sim_matrix.max():.4f}]")

for _, row in tqdm(test_pairs_df.iterrows(), total=len(test_pairs_df), desc="Computing similarities"):
    query_id = img_to_index[row['query_image']]
    gallery_id = img_to_index[row['gallery_image']]
    similarity = sim_matrix[query_id, gallery_id]
    similarities.append(similarity)

# Clip to [0, 1] range
similarities = np.array(similarities)
min_sim, max_sim = similarities.min(), similarities.max()
similarities = (similarities - min_sim) / (max_sim - min_sim + 1e-8)
# similarities = np.clip(similarities, 0.0, 1.0)

print(f"\nSimilarity statistics:")
print(f"  Min: {similarities.min():.4f}")
print(f"  Max: {similarities.max():.4f}")
print(f"  Mean: {similarities.mean():.4f}")
print(f"  Std: {similarities.std():.4f}")
```
